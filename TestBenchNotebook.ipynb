```python
import RunDjikstra as D
import time
import os
path_info =[]
djikstra_run=[]
djikstra_run_times=[]
image_sequence=[]

shortest_path=[]
nodes_visited=[]
```

```python

directory='Sample_Mazes/LoopMazes'
for filename in os.listdir(directory):
    if filename.endswith(".png"): 
        image_file=os.path.join(directory, filename)
        image_sequence.append(filename)
        start_time = time.time()
        path_info=D.run_djikstra(image_file)
        djikstra_run_times.append(time.time() - start_time) 
        djikstra_run.append(path_info)
        print('Done with: '+image_file)
        
    else:
        continue

```

```python
for i in range(len(djikstra_run)):
    shortest_path.append(djikstra_run[i][0])
    nodes_visited.append(djikstra_run[i][1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

img_size=[]
img_size.append(84100)
img_size.append(103684)
img_size.append(9604)
img_size.append(58564)
img_size.append(6724)
img_size.append(16900)
img_size.append(37636)
img_size.append(21316)
img_size.append(26244)

img_size=np.asarray(img_size)
print(type(img_size))
djikstra_run_times=np.asarray(djikstra_run_times)
```

```python
import RunAstar as astar

astar_path_info =[]
astar_run=[]
astar_run_times=[]
astar_image_sequence=[]

astar_shortest_path=[]
astar_nodes_visited=[]

directory='Sample_Mazes/LoopMazes'
for filename in os.listdir(directory):
    if filename.endswith(".png"): 
        image_file=os.path.join(directory, filename)
        astar_image_sequence.append(filename)
        start_time = time.time()
        path_info=astar.runAstar(image_file)
        astar_run_times.append(time.time() - start_time) 
        astar_run.append(path_info)
        print('Done with: '+image_file)
        
    else:
        continue
```

```python
for i in range(len(astar_run)):
    astar_shortest_path.append(astar_run[i][0])
    astar_nodes_visited.append(astar_run[i][1])
```

```python
import RunBfs as BFS
bfs_path_info =[]
bfs_run=[]
bfs_run_times=[]
bfs_image_sequence=[]

bfs_shortest_path=[]
bfs_nodes_visited=[]

directory='Sample_Mazes/LoopMazes'
for filename in os.listdir(directory):
    if filename.endswith(".png"): 
        image_file=os.path.join(directory, filename)
        bfs_image_sequence.append(filename)
        start_time = time.time()
        path_info=BFS.run_BFS(image_file)
        bfs_run_times.append(time.time() - start_time) 
        bfs_run.append(path_info)
        print('Done with: '+image_file)
        
    else:
        continue
```

```python
for i in range(len(bfs_run)):
    bfs_shortest_path.append(astar_run[i][0])
    bfs_nodes_visited.append(astar_run[i][1])
```

```python
plt.figure(figsize=(14,6))
plt.plot(sorted(img_size),sorted(djikstra_run_times))
plt.plot(sorted(img_size),sorted(astar_run_times))
plt.plot(sorted(img_size),sorted(bfs_run_times))
plt.legend(("Djikstra Algorithm", "A* Algorithm","BFS Algorithm"))
plt.xlabel("Image Size (HeightXWidth)")
plt.ylabel("Computational time (s)")
plt.title('Comparitive Plot of runtime against image size')
plt.show()
```

```python
# List of distance costs using Nearest Neighbor and 2-opt from random paths
#costs_comp = [[best_dist_10,best_dist_20,best_dist_100],[cost_10_2opt,cost_20_2opt,cost_100_2opt]]
X=np.arange(9)
bar_width = 0.25

# Visualizing the results
fig=plt.figure(figsize=(14,6))
ax = fig.add_axes([0,0,1,1])
rects1 = plt.bar(X - bar_width, shortest_path, bar_width, color='b', label='Djikstra Algorithm')
rects2 = plt.bar(X, astar_shortest_path, bar_width, color='g', label='A*')
rects3 = plt.bar(X+ bar_width , bfs_shortest_path, bar_width, color='r', label='BFS')
plt.xlabel('Image Size')
plt.ylabel('Shortest Path')
plt.title('Comparative plot Shortest Path found per algorithm against Image Sizes')
plt.xticks(X + bar_width/2, img_size)
plt.yscale("log")
plt.legend()
plt.show()
```
